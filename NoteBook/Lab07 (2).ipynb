```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('ventas-por-factura.csv')
```

```python
df.head
```

```python
plt.figure(figsize=(10, 5))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('2. Heatmap de Valores Nulos en el Dataset')
plt.show()
```

```python
plt.figure(figsize=(10, 4))
try:
    sns.boxplot(data=df, x='Monto', color='salmon')
    plt.title('3. Distribución del Monto (Detección de Outliers y Negativos)')
except Exception:
    pass
plt.show()
```
